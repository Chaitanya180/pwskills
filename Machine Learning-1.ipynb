
### Q1

#### 1) Artificial Intelligence
Artificial Intelligence (AI) refers to the development of computer systems that can perform tasks that typically require human intelligence. This includes tasks such as visual perception, speech recognition, decision-making, and language translation. AI can be classified into two types: Narrow or Weak AI, which is designed for a particular task, and General or Strong AI, which aims to perform any intellectual task that a human being can do.

Example: Siri or Alexa, virtual assistants that understand and respond to voice commands.

#### 2) Machine Learning
Machine Learning (ML) is a subset of AI that involves the development of algorithms that enable computers to learn from data. ML systems can improve their performance on a specific task over time without being explicitly programmed.

Example: Spam email filters that learn to distinguish between spam and non-spam emails based on user feedback.

#### 3) Deep Learning
Deep Learning (DL) is a subfield of machine learning that deals with neural networks containing multiple layers (deep neural networks). DL has been particularly successful in tasks such as image and speech recognition.

Example: Image recognition systems using convolutional neural networks (CNNs).

### Q2
#### Supervised Learning
Supervised learning is a type of machine learning where the algorithm is trained on a labeled dataset, meaning that the input data is paired with the corresponding correct output. The goal is for the algorithm to learn a mapping from inputs to outputs.

Examples:
1. Handwriting recognition: Training a model to recognize handwritten characters with labeled examples.
2. Email spam detection: Training a model to classify emails as spam or not spam based on labeled training data.

### Q3
#### Unsupervised Learning
Unsupervised learning involves training a model on an unlabeled dataset, and the algorithm must find patterns and relationships in the data without explicit guidance on the output.

Examples:
1. Clustering: Grouping similar data points together, such as customer segmentation based on purchase behavior.
2. Dimensionality reduction: Reducing the number of features in a dataset while preserving its essential information.

### Q4
#### Differences between AI, ML, DL, and DS
- AI (Artificial Intelligence): The broader concept of creating machines that can perform tasks that typically require human intelligence.
- ML (Machine Learning): A subset of AI that focuses on developing algorithms allowing computers to learn from data.
- DL (Deep Learning): A subfield of ML that deals with neural networks containing multiple layers.
- DS (Data Science): Involves extracting insights and knowledge from structured and unstructured data, utilizing various techniques, including machine learning.

### Q5
#### Differences between Supervised, Unsupervised, and Semi-Supervised Learning
- Supervised Learning: Requires labeled training data with input-output pairs.
- Unsupervised Learning: Works with unlabeled data, discovering patterns and relationships.
- Semi-Supervised Learning: Combines labeled and unlabeled data for training.

### Q6
#### Train, Test, and Validation Split
- Train Set: Data used to train the model.
- Test Set: Independent dataset used to evaluate the model's performance.
- Validation Set: Subset of the training data used to fine-tune the model parameters.

Importance:
- Train Set: Used to train the model and adjust weights.
- Validation Set: Helps prevent overfitting by tuning hyperparameters during training.
- Test Set: Evaluates the model's performance on new, unseen data.

### Q7
#### Unsupervised Learning in Anomaly Detection
Unsupervised learning can be used in anomaly detection by identifying patterns in data and flagging instances that deviate significantly from these patterns.

Example: Anomaly detection in network security, identifying unusual patterns that may indicate a cyber attack.
